```python
#Import
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('headbrain.csv')
```

```python
#printing row and column
print(dataset.shape)
```

```python
dataset
X=dataset['Head Size(cm^3)'].values
Y=dataset['Brain Weight(grams)'].values
```

```python
#Mean of X and Y
meanx=np.mean(X)
meany=np.mean(Y)
n=len(X)
```

```python
#Calculate m and c using formula (y=mx+c)
sum1=0
sum2=0
for i in range(n):
    sum1+=(X[i]-meanx)*(Y[i]-meany)
    sum2+=(X[i]-meanx)**2
m=sum1/sum2
c=meany-(m*meanx)
print(m,c)
```

```python
minx=np.min(X)
maxx=np.max(X)
x=np.linspace(minx,maxx,1000)
y=m*x+c
#Plotting Line
plt.plot(x,y,color='#e01a10',label='Regression Line')
#Plotting scatter point
plt.scatter(X,Y,c='#288a79',label='Scatter Plot')
plt.xlabel('Head Size(cm^3)')
plt.ylabel('Brain Weight(grams)')
plt.show()
```

```python
#R square Value
ss1=0
ss2=0
for i in range(n):
    ypred=c+m*X[i]
    ss1+=(Y[i]-meany)**2
    ss2+=(Y[i]-ypred)**2
R2=1-(ss2/ss1)
R2
```
